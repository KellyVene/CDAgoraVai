## 2015

```python
df = pd.read_csv('DOM2015.csv',sep=',')
```

```python
#Trabalhar com todos os domicílios considerados permanente V0201 = 1
#Trabalhar com todos os domicílios Próprio Pago (1), Próprio Pagando (2), Alugado (3) 

df = df.dropna(subset=['V4621'])
df = df[df.V4621<3000000]
df = df[df.V0207<=3]
```

```python
df["REGIAO"]='x'
df.loc[(df['UF'] >=10) & (df['UF'] < 20),'REGIAO'] = 'Norte'
df.loc[(df['UF'] >=20) & (df['UF'] < 30),'REGIAO'] = 'Nordeste'
df.loc[(df['UF'] >=30) & (df['UF'] < 40),'REGIAO'] = 'Sudeste'
df.loc[(df['UF'] >=40) & (df['UF'] < 50),'REGIAO'] = 'Sul'
df.loc[(df['UF'] >=50) & (df['UF'] < 60),'REGIAO'] = 'Centro-Oeste'
```

```python
#V0231 - Tem microcomputador
df.V0231 = df.V0231.astype('category')
df.V0231.cat.categories = ('Sim','Não')

#V02322 - Tem acesso a net pelo próprio domicílio
df.V02322 = df.V02322.astype('category')
df.V02322.cat.categories = ('Sim','Não')
```

## 2013

```python
dt = pd.read_csv('DOM2013.csv',sep=',')
```

```python
#Trabalhar com todos os domicílios considerados permanente V0201 = 1
#Trabalhar com todos os domicílios Próprio Pago (1), Próprio Pagando (2), Alugado (3) 

dt = dt.dropna(subset=['V4621'])
dt = dt[dt.V4621<3000000]
dt = dt[dt.V0207<=3]
```

```python
dt["REGIAO"]='x'
dt.loc[(dt['UF'] >=10) & (dt['UF'] < 20),'REGIAO'] = 'Norte'
dt.loc[(dt['UF'] >=20) & (dt['UF'] < 30),'REGIAO'] = 'Nordeste'
dt.loc[(dt['UF'] >=30) & (dt['UF'] < 40),'REGIAO'] = 'Sudeste'
dt.loc[(dt['UF'] >=40) & (dt['UF'] < 50),'REGIAO'] = 'Sul'
dt.loc[(dt['UF'] >=50) & (dt['UF'] < 60),'REGIAO'] = 'Centro-Oeste'
```

```python
#V0231 - Tem microcomputador
dt.V0231 = dt.V0231.astype('category')
dt.V0231.cat.categories = ('Sim','Não')

#V02322 - Tem acesso a net pelo próprio domicílio
dt.V02322 = dt.V02322.astype('category')
dt.V02322.cat.categories = ('Sim','Não')
```

## Análise de duas variáveis qualitativas por região usando comando `.reindex`

### Análise correta!!!!

```python
vars = ('V0231','V02322')
nomevar = (
         'V0231 - Tem microcomputador?',
         'V02322 - Tem acesso a net pelo próprio domicílio')

for var in vars:
    print(' ')    
    print(' ')
    print(nomevar[vars.index(var)])
    print(' ')
    print('2015')
    
    fig = plt.figure(figsize=(15, 3))

    plt.subplot(151)
    a = df[var][df.REGIAO == 'Norte'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Norte')
    plt.ylabel(' ')

    plt.subplot(152)
    a = df[var][df.REGIAO == 'Nordeste'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Nordeste')
    plt.ylabel(' ')

    plt.subplot(153)
    a = df[var][df.REGIAO == 'Sul'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Sul')
    plt.ylabel(' ')

    plt.subplot(154)
    a = df[var][df.REGIAO == 'Sudeste'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Sudeste')
    plt.ylabel(' ')

    plt.subplot(155)
    a = df[var][df.REGIAO == 'Centro-Oeste'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Centro-Oeste')
    plt.ylabel(' ')

    plt.tight_layout()
    plt.show()
    
    print('2013')
    
    fig = plt.figure(figsize=(15, 3))

    plt.subplot(151)
    a = dt[var][dt.REGIAO == 'Norte'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Norte')
    plt.ylabel(' ')

    plt.subplot(152)
    a = dt[var][dt.REGIAO == 'Nordeste'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Nordeste')
    plt.ylabel(' ')

    plt.subplot(153)
    a = dt[var][dt.REGIAO == 'Sul'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Sul')
    plt.ylabel(' ')

    plt.subplot(154)
    a = dt[var][dt.REGIAO == 'Sudeste'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Sudeste')
    plt.ylabel(' ')

    plt.subplot(155)
    a = dt[var][dt.REGIAO == 'Centro-Oeste'].value_counts().reindex(['Sim','Não'])
    a.plot(kind='pie', autopct="%.2f",legend=False)
    plt.title('Centro-Oeste')
    plt.ylabel(' ')

    plt.tight_layout()
    plt.show()
```
